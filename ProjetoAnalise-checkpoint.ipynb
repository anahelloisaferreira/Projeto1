```python
#Projeção de um determinado limite aceitável baseado na curva Teórica
#Através do dataframe criamos uma nova coluna que define o que está dentro dos limites aceitáveis ou não

potencia_real = turbina ["Potência(KW)"].tolist()
potencia_teorica = turbina ["CurvaTeorica(KWh)"].tolist()

#Criação de lista que será de potência máxima e de potência mínima
#Potência máxima será o valor da turbina teórica + 5%
#Potêncima mínima será o valor da potência teórica - 5%
#Baseado na potência teórica criamos um limite superior e inferior

potencia_max = []
potencia_min = []
dentro_limite = []

for potencia in potencia_teorica:
    potencia_max.append(potencia*1.05)
    potencia_min.append(potencia*0.95)
    
#Comparativo das potências, verificando se os dados estão dentro dos limites propostos acima
#Através do limite iremos perceber quem estará dentro ou fora dos limites
#Após analisar os limites podemos perceber os valores que estão fora desses limites

for p, potencia in enumerate(potencia_real):
    if potencia>=potencia_min[p] and potencia<=potencia_max[p]:
        dentro_limite.append("Dentro")
    elif potencia==0:
        dentro_limite.append("Zero")
    else:
        dentro_limite.append("Fora")

#Se caso os dados não forem verdadeiros os dados que não estão dentro do limite da curva teórica vão para fora
#Potência dentro dos limistes, potência zerada ou não estão em nenhuma das duas são considerados fora

print(dentro_limite.count("Dentro")/len(dentro_limite))

#Mostrando quais valores estão dentro da minha lista de limites
```
